```python
import os
from ultralytics import YOLO
import cv2

# Load the YOLOv8 model
model = YOLO('final.pt')  # Specify the correct path to your .pt file
image="test.jpg"
# Perform detection
results = model.predict(source=image, conf=0.8, save=False)
img = cv2.imread(image)
# Create the 'nails' folder if it doesn't exist
output_folder = "nails"
os.makedirs(output_folder, exist_ok=True)



# Iterate over the detected results
nail_count = 0
for result in results:
    boxes = result.boxes  # Get the detected boxes
    for box in boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates

        # Crop the detected nail from the original image
        cropped_nail = img[y1:y2, x1:x2]

        # Save the cropped nail image to the 'nails' folder
        nail_count += 1
        output_path = os.path.join(output_folder, f"nail_{nail_count}.jpeg")
        cv2.imwrite(output_path, cropped_nail)

        print(f"Saved: {output_path}")

#display one of the cropped images (for example, the first one) in Jupyter Notebook
if nail_count > 0:
    cropped_nail_rgb = cv2.cvtColor(cropped_nail, cv2.COLOR_BGR2RGB) 
    import matplotlib.pyplot as plt
    plt.figure(figsize=(6, 6))
    plt.imshow(cropped_nail_rgb)
    plt.axis('off')
    plt.show()
```
